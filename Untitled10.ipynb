```python
from sklearn.gaussian_process import GaussianProcessRegressor

#     
model = lambda x:x*np.sin(x)
xdata = np.array([1, 3, 5, 6, 8, 10])
a = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

print(a.reshape(-1, 1), ydata.reshape(-1, 1))

#    
gp = GaussianProcessRegressor()
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 30, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], return_std=True)
dyfit=2*np.sqrt(MSE) #2* ~     95%

plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit-dyfit, yfit+dyfit, color='gray', alpha=0.2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np

def f(x, y):
    return np.sin(x)**10 + np.cos(10+ y* x)*np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5 ,40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contour(X, Y, Z, colors='black')
```

```python
plt.contourf(X, Y, Z,20, cmap='RdGy')
plt.colorbar()
```

```python
plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy')
plt.colorbar()
```

```python
contours = plt.contour(X,Y,Z, 4, colors='black')
plt.clabel(contours, inline=True, fontsize=15)
plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('classic')

x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-r', label='Sine')
ax.plot(x, np.cos(x), '--b', label='Cosine')
ax.axis('equal')
ax.legend()

```
